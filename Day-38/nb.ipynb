```python
def merge_with_teams_data(battle_data:pd.DataFrame,player_battle_log:pd.DataFrame):
    left_cols = list(battle_data.columns)
    left_cols.remove("teams_data")
    right_cols = list(player_battle_log.columns)
    resulting_cols = left_cols+right_cols
    merged_data_frame = pd.DataFrame(columns=resulting_cols)
    for i in range(len(battle_data)):
        keys = battle_data.loc[i,"teams_data"][0].keys()
        for key in keys:
            for player in battle_data.loc[i,"teams_data"][0][key]:
                idx = len(merged_data_frame)
                # print(type(battle_data.loc[i,left_cols]))
                # print(pd.concat([battle_data.loc[i,left_cols],player_battle_log[player_battle_log["playerMatchId"]==player]],axis=1).columns)
                # print(merged_data_frame.columns)
                merged_data_frame.loc[idx] = battle_data.loc[i,left_cols]
                for col in right_cols:
                    merged_data_frame.loc[idx,col] =  (player_battle_log[player_battle_log["playerMatchId"]==player][col].values[0])
    
    return merged_data_frame

```
